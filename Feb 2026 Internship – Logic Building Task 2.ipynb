```python
#1. question
sentence=input("Enter the sentence:")
word=sentence.split()
unique_word=set(word)
print("Length of unique_word: ",len(unique_word))
print("Unique words are: ",unique_word)
```

```python
#2. question
employees={"Ravi": 75000,"Anita": 68000,"Kiran": 72000}
highest_employee=max(employees,key=employees.get)
highest_salary=employees[highest_employee]
print(f"Highest Salary:{highest_employee}-{highest_salary}")
```

```python
#3. question
numbers=[45,22,89,10,66]
maximum=numbers[0]
minimum=numbers[0]
for num in numbers:
    if num>maximum:
        maximum=num
    if num<minimum:
        minimum=num
print("List:",numbers)
print("Max:",maximum)
print("Min:",minimum)
```

```python
#4. question
prices=[450,1200,899,1500,300]
cnt=0
for price in prices:
    if price>1000:
        cnt+=1
print("Products above 1000 count are:",cnt)
```

```python
#5. question
attendance=["P","P","A","P","P"]
present_count=0
for status in attendance:
    if status=="P":
        present_count+=1
total_days=len(attendance)
percentage=(present_count/total_days)*100
print("Attendance Percentage:",percentage)
```

```python
#7. question
text="python"
char_count={}
for ch in text:
    if ch in char_count:
        char_count[ch]+=1
    else:
        char_count[ch]=1
print(char_count)
```
