```python
#1. question
def social_media(like):
  tot=0
  for i in like:
    tot+=i
  if tot>=1000:
    status="viral post"
  else:
    status="Normal Engagement"
  print("Total likes: ",tot)
  print("Status: ",status)
like=[100,200,150,122,300]
social_media(like)
```

```python
#2. question
def medicine(stock):
    if stock<10:
        status="Low Stock Alert"
    else:
        status="Stock Sufficient"
    print("Medicine Stock:", stock)
    print("Status:", status)
medicine_stock=6
medicine(medicine_stock)
```

```python
#3. question
def check_rainfall(rainfall,required_level):
    tot=0
    for value in rainfall:
        tot+=value
    average=tot/len(rainfall)
    if average>=required_level:
        status="Adequate Rainfall"
    else:
        status="Inadequate Rainfall"
    print("Average Rainfall:",int(average))
    print("Rainfall Status:",status)
rainfall_data=[60,80,75,70,75]
required=70
check_rainfall(rainfall_data,required)
```

```python
#4. question
def check_duplicates(usernames):
    if len(usernames)!=len(set(usernames)):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")
users=["user1","user2","user3","user1"]
check_duplicates(users)
```

```python
#6. question
def filter_premium_crops(prices):
    premium= []
    for price in prices:
        if price>2000:
            premium.append(price)
    print("Premium Crops:",premium)
crop_prices=[1800,2500,1500,3200,2000]
filter_premium_crops(crop_prices)
```

```python
#7. question
def check_system_health(errors):
    if errors==0:
        status="Healthy"
    elif errors<=5:
        status="Minor Issues"
    else:
        status="Critical Issues"
    print("Error Count:",errors)
    print("System Status:",status)
error_count=7
check_system_health(error_count)
```

```python
#9. question
def check_exam_eligibility(attendance):
    present=0
    for day in attendance:
        if day=="P":
            present+=1
    percentage=(present/len(attendance))*100
    if percentage>=75:
        status="Eligible"
    else:
        status="Not Eligible"
    print("Attendance Percentage:",percentage)
    print("Exam Eligibility:",status)
attendance_list=["P","P","A","P","P"]
check_exam_eligibility(attendance_list)
```

```python
#10. question
def calculate_bill(units):
    bill=0

    if units<=100:
        bill=units*3
    elif units<=200:
        bill=(100*3)+((units-100)*5)
    else:
        bill=(100*3)+(100*5)+((units-200)*7)

    if bill<500:
        status="Low Usage"
    elif bill<=1500:
        status="Moderate Usage"
    else:
        status="High Usage"
    return bill,status
units_consumed=250

total_bill,usage_status=calculate_bill(units_consumed)
print("Units Consumed:",units_consumed)
print("Total Bill: ₹",total_bill)
print("Usage Level:",usage_status)
```
